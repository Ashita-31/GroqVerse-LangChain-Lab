### What is langchain?
#### it is an open source frame work design to simplify the creation of application using LLM (large language model).

## 1.Technical Defination of a framework 
##### a framework is a universal, reusable environment that provides specific functionality as part of a larger software platform to facilitate the development of software applications, products, and solutions. 


### 2.Technical defination of LLM (Large Language Model)
##### An LLM is an autoregressive transformer-based neural network that uses self-attention mechanisms to process high-dimensional vector representations of text. It functions as a probabilistic engine, calculating a likelihood distribution over a fixed vocabulary to predict the most statistically probable next token.

```python
import os
import getpass
from langchain_groq import ChatGroq
```

```python
os.environ["GROQ_API_KEY"] = getpass.getpass("Enter your Groq API key: ")
```

```python
# 2. Initialize the Groq model (Llama 3.3 is currently a top free choice)
groq_chat = ChatGroq(
    model="llama-3.3-70b-versatile",
    temperature=0
)
```

```python
# 3. Simple Test
response = groq_chat.invoke("Explain ML in one sentence.")
print(f"Groq Response: {response.content}")
print("✅ Groq connector is ready!")
     
```
